```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import signal

def wiener_filter(data, template):
    data_fft = np.fft.fft(data)
    template_fft = np.fft.fft(template)
    data_fft_freqs = np.fft.fftshift(np.fft.fftfreq(data.size))
    template_fft_freqs = np.fft.fftshift(np.fft.fftfreq(template.size))

    w_filter = np.abs(np.fft.fft(data_fft * template_fft))
        
    return w_filter

def lorentzian(x, center_freq, gamma, a):
    return a * 0.5 * gamma**2 / (0.5 * gamma**2 + (x - center_freq)**2)

def lorentzify(signal, gamma, a=1/np.pi, center_freq=0):
    signal_fft = np.fft.fft(signal)
    signal_fft_freqs = np.fft.fftshift(np.fft.fftfreq(signal.size))
    print("pre lorentz: ")
    timeseries_plot(signal_fft, freq=signal_fft_freqs, xlabel='freq', ylabel='power', title='FFT')
    signal_fft = signal_fft * lorentzian(signal_fft_freqs, center_freq, gamma, a)
    print("lorentzified: ")
    timeseries_plot(signal_fft, freq=signal_fft_freqs, xlabel='freq', ylabel='power', title='FFT')
    return np.fft.ifft(signal_fft)

def timeseries_plot(data, freq=None, xlabel='x', ylabel='y', title=None):
    timesteps = np.arange(len(data))
    if freq is not None:
        timesteps = freq
    plt.plot(timesteps, data)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    if title is not None:
        plt.title(title)
    plt.show()
```

```python
# Init params
noise_length = 30000
start_time = 12800
alpha = 0.0001
amplitude = 2
gamma = 0.1
temp_alpha = 0.0001
temp_amplitude = 2

# Generate white noise
print("----- method 1 : lorenzify noise and signal separately -----")
rng = np.random.default_rng()
white_noise = rng.standard_normal(noise_length)
print("white noise plot: ")
timeseries_plot(white_noise, xlabel='time', ylabel='amplitude', title='timeseries')
lorentz_white_noise = lorentzify(white_noise, gamma)
print("lorentzified white noise plot: ")
timeseries_plot(lorentz_white_noise, xlabel='time', ylabel='amplitude', title='timeseries')

# Generate a signal
sig_length = noise_length - start_time
sig_timesteps = np.arange(sig_length)
signal = amplitude * np.sin(alpha * np.square(sig_timesteps))
print("signal plot: ")
timeseries_plot(signal, xlabel='time', ylabel='amplitude', title='timeseries')
lorentz_signal = lorentzify(signal, gamma)
print("lorentzified signal plot: ")
timeseries_plot(lorentz_signal, xlabel='time', ylabel='amplitude', title='timeseries')

# Inject the signal
left_noise = white_noise[:start_time]
left_zeros = np.zeros(len(left_noise))

# Lorenzify signal and noise separately and combine
extended_lorentz_signal = np.concatenate((left_zeros, lorentz_signal), axis=None)
lorentz_sig_noise = lorentz_white_noise + extended_lorentz_signal
print("combined lorentz plot: ")
timeseries_plot(lorentz_sig_noise, xlabel='time', ylabel='amplitude', title='timeseries')

# Combine signal and noise and then lorentzify
print("----- method 2 : combine noise and signal first then lorentzify -----")
extended_sig = np.concatenate((left_zeros, signal), axis=None)
sig_noise = white_noise + extended_sig
print("white noise + injected signal plot: ")
timeseries_plot(sig_noise, xlabel='time', ylabel='amplitude', title='timeseries')
lorentz_noise = lorentzify(sig_noise, gamma)
print("lorentz_noise plot: ")
timeseries_plot(lorentz_noise, xlabel='time', ylabel='amplitude', title='timeseries')

# Spectrogram the signal
print("----- spectrogram the signal -----")
powerSpectrum, freqenciesFound, time, imageAxis = plt.specgram(lorentz_noise, Fs=6000)
plt.xlabel('Time')
plt.ylabel('Frequency')
print("matplotlib.spectrogram plot (white noise + signal): ")
plt.show()

# Generate a signal template
print("----- generate a signal template -----")
template_length = noise_length
template_timesteps = np.arange(template_length)
template = temp_amplitude * np.sin(temp_alpha * np.square(template_timesteps))
print("template plot (alpha = 0.0001): ")
timeseries_plot(template, xlabel='time', ylabel='amplitude', title='timeseries')
lorentz_template = lorentzify(template, gamma)
print("lorentz_template plot: ")
timeseries_plot(lorentz_template, xlabel='time', ylabel='amplitude', title='timeseries')

# Wiener filter
print("----- use wiener filter -----")
filtered = wiener_filter(lorentz_noise, lorentz_template)
print("filter plot (template alpha = 0.00010):")
timeseries_plot(filtered)
```
